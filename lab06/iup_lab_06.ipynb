```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Image filtering using 2D convolution

# Load the image
noise_img1 = cv.imread("D:\\SLIIT\\4th_year\\2nd semester\\IUP\LABS\\lab06\\saltandpepper1.png", 1)
noise_img2 = cv.imread("D:\\SLIIT\\4th_year\\2nd semester\\IUP\LABS\\lab06\\saltandpepper2.png", 1)

# Apply custom averaging filter (Lowpass Filter) using filter2D()
kernal = np.ones((5, 5), np.float32)/25
dst1 = cv.filter2D(noise_img1, -1, kernal)
dst2 = cv.filter2D(noise_img2, -1, kernal)

result1 = np.hstack((noise_img1, dst1))
result2 = np.hstack((noise_img2, dst2))
cv.imshow("Filtered Image", result1)
cv.waitKey();
plt.imshow(cv.cvtColor(result1, cv.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv.cvtColor(result2, cv.COLOR_BGR2RGB))
plt.show()
```

```python
# 2. Image Averaging using Box Filter

# Load the image
noise_img1 = cv.imread("D:\\SLIIT\\4th_year\\2nd semester\\IUP\LABS\\lab06\\saltandpepper1.png", 1)
noise_img2 = cv.imread("D:\\SLIIT\\4th_year\\2nd semester\\IUP\LABS\\lab06\\saltandpepper2.png", 1)

# Apply averaging filter (Lowpass Filter) : (3 * 3 filter)
dst1 = cv.blur(noise_img1, (3, 3))
dst2 = cv.blur(noise_img2, (3, 3))

result1 = np.hstack((noise_img1, dst1))
result2 = np.hstack((noise_img2, dst2))
cv.imshow("Filtered Image", result1)
cv.waitKey();
plt.imshow(cv.cvtColor(result1, cv.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv.cvtColor(result2, cv.COLOR_BGR2RGB))
plt.show()
```

```python
# 2. Image Averaging using Box Filter change avg

# Load the image
noise_img1 = cv.imread("D:\\SLIIT\\4th_year\\2nd semester\\IUP\LABS\\lab06\\saltandpepper1.png", 1)
noise_img2 = cv.imread("D:\\SLIIT\\4th_year\\2nd semester\\IUP\LABS\\lab06\\saltandpepper2.png", 1)

# Apply averaging filter (Lowpass Filter) : (3 * 3 filter)
dst1 = cv.boxFilter(noise_img1, -1, (3, 3))
dst2 = cv.boxFilter(noise_img2, -1, (3, 3))

result1 = np.hstack((noise_img1, dst1))
result2 = np.hstack((noise_img2, dst2))
cv.imshow("Filtered Image", result1)
cv.waitKey();
plt.imshow(cv.cvtColor(result1, cv.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv.cvtColor(result2, cv.COLOR_BGR2RGB))
plt.show()
```

```python
# Median Filtering

# Load the image
noise_img = cv.imread("D:\\SLIIT\\4th_year\\2nd semester\\IUP\LABS\\lab06\\saltandpepper1.png", 1)

# Apply averaging filter : (Median value = 3)
dst = cv.medianBlur(noise_img, 3)

result = np.hstack((noise_img, dst))
# cv.imshow("Filtered Image", result)
# cv.waitKey();
plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB))
plt.show()
```

```python
# Gaussian Filtering

# Load the image
noise_img = cv.imread("D:\\SLIIT\\4th_year\\2nd semester\\IUP\LABS\\lab06\\gaussian Noise.png", 1)

# Apply averaging filter : (Median value = 3)
dst = cv.GaussianBlur(noise_img, (11, 11), 0)

result = np.hstack((noise_img, dst))
# cv.imshow("Filtered Image", result)
# cv.waitKey();
plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB))
plt.show()
```
