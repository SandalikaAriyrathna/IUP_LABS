```python
import cv2 as cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Histogram Equalization to improve the contrast of a dark image

dark_img = cv2.imread("D:\\SLIIT\\4th_year\\2nd semester\\IUP\LABS\\lab04\\darkImage.png", 0)

# Create histogram of the image
dark_img_hist = cv2.calcHist([dark_img], [0], None, [256], [0, 256])

# Apply histogram equalization
equalized_dark_img = cv2.equalizeHist(dark_img)
# Create histogram of the equalized image
equalized_dark_img_hist = cv2.calcHist([equalized_dark_img], [0], None, [256], [0, 256])

# Plot the dark_img_hist & equalized_dark_img_hist
# : plt.subplot(211) - 1st parameter value means Row, 2nd parameter value means column, 3rd parameter value means image no
plot1 = plt.subplot(211), plt.plot(dark_img_hist)
plot2 = plt.subplot(212), plt.plot(equalized_dark_img_hist)
plt.show(plot1)
plt.show(plot2)

# Equalized Image window will be appered firstly, After closing that window histogram graphs will be displyed
result = np.hstack((dark_img, equalized_dark_img))
cv2.imshow("Equalized Image", result)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.imshow(result)
```

```python
# Histogram Equalization to improve the contrast of a dark and bright image-obtain the outputs below

dark_img = cv2.imread("D:\\SLIIT\\4th_year\\2nd semester\\IUP\LABS\\lab04\\darkImage.png", 0)
bright_img = cv2.imread("D:\\SLIIT\\4th_year\\2nd semester\\IUP\LABS\\lab04\\brightImage.png", 0)

# Create histogram of the image
dark_img_hist = cv2.calcHist([dark_img], [0], None, [256], [0, 256])
bright_img_hist = cv2.calcHist([bright_img], [0], None, [256], [0, 256])

# Apply histogram equalization
equalized_dark_img = cv2.equalizeHist(dark_img)
equalized_bright_img = cv2.equalizeHist(bright_img)
# Create histogram of the equalized image
equalized_dark_img_hist = cv2.calcHist([equalized_dark_img], [0], None, [256], [0, 256])
equalized_bright_img_hist = cv2.calcHist([equalized_bright_img], [0], None, [256], [0, 256])

# Plot the dark_img_hist & equalized_dark_img_hist
# : plt.subplot(211) - 1st parameter value means Row, 2nd parameter value means column, 3rd parameter value means image no
plt.subplot(221), plt.plot(dark_img_hist)
plt.subplot(222), plt.plot(bright_img_hist)
plt.subplot(223), plt.plot(equalized_dark_img_hist)
plt.subplot(224), plt.plot(equalized_bright_img_hist)
```

```python
# Color Histogram Equalization to improve the contrast of color image

color_img = cv2.imread("D:\\SLIIT\\4th_year\\2nd semester\\IUP\LABS\\lab04\\color.png", 1)

# Convert image from RGB to HSV
color_img_hsv = cv2.cvtColor(color_img, cv2.COLOR_RGB2HSV)

# Histogram equalization on V channel
color_img_hsv[:,:,2] = cv2.equalizeHist(color_img_hsv[:,:,2])

# Convert image from HSV to RGB
color_img_rgb = cv2.cvtColor(color_img_hsv, cv2.COLOR_HSV2RGB)

# Create histogram of the image
color_img_hist = cv2.calcHist([color_img], [0], None, [256], [0, 256])
color_img_rgb_hist = cv2.calcHist([color_img_rgb], [0], None, [256], [0, 256])

# Plot the color_img_hist & equalized_color_img_hist
# : plt.subplot(211) - 1st parameter value means Row, 2nd parameter value means column, 3rd parameter value means image no
plt.subplot(211), plt.plot(color_img_hist)
plot2 = plt.subplot(212), plt.plot(color_img_rgb_hist)
plt.show(plot2)

# Equalized Image window will be appered firstly, After closing that window histogram graphs will be displyed
result = np.hstack((color_img, color_img_rgb))
cv2.imshow("Equalized Image", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(result)
```
