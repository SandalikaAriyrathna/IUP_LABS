```python
import cv2
import numpy as np
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
```

```python
# convert images from one color-space to another,
# like BGR ↔ Gray, BGR ↔ HSV,
img = cv2.imread("D:\\SLIIT\\4th_year\\2nd semester\\IUP\LABS\\lab02\\sunflower.jpg", 1)
cv2.imshow('Original Image',img)

##convert bgr to gray
imgray =cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) 
cv2.imshow('Gray Image',imgray)

#convert to hsv 
imghsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV) 
cv2.imshow('HSV IMAGE',imghsv)

# convert to binary image
res, binary = cv2.threshold(imgray, 120, 255, cv2.THRESH_BINARY)
cv2.imshow('Binary Image', binary)

cv2.waitKey() 
cv2.destroyAllWindows()
```

```python
fig,ax = plt.subplots(1,4, figsize = (15, 15))
ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax[0].axis('on')
ax[1].imshow(cv2.cvtColor(imgray, cv2.COLOR_BGR2RGB))
ax[1].axis('off')
ax[2].imshow(cv2.cvtColor(imghsv, cv2.COLOR_BGR2RGB))
ax[2].axis('on')
ax[3].imshow(cv2.cvtColor(binary, cv2.COLOR_BGR2RGB))
ax[3].axis('on')
```

```python
# Threshold the HSV image for a range of blue color

import cv2
import numpy as np

# Load the video
video = cv2.VideoCapture("D:\\SLIIT\\4th_year\\2nd semester\\IUP\\LABS\\lab02\\small_laptop_connections.mov")

# Read frame by frame
while video.isOpened():
    ret, frame = video.read()
    
    # Check if the frame was read successfully
    if not ret:
        break  # Exit the loop if no frame is read

    # Convert BGR to HSV color space
    videoHSV = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    # Define the upper and lower ranges for the blue color
    lower_blue = np.array([110, 50, 50])
    upper_blue = np.array([130, 255, 255])
    
    # Threshold the HSV image to get only blue colors
    maskVideo = cv2.inRange(videoHSV, lower_blue, upper_blue)
    
    # Resize the masked video
    originalVideo = cv2.resize(maskVideo, (400, 400))
    cv2.imshow('OriginalVideo', originalVideo)
    
    # Break loop on 'q' key press
    if cv2.waitKey(1) == ord('q'):
        break
    
video.release()
cv2.destroyAllWindows()
```

```python
# Now extract the blue object alone

import cv2
import numpy as np

# Load the video
video = cv2.VideoCapture("D:\\SLIIT\\4th_year\\2nd semester\\IUP\\LABS\\lab02\\small_laptop_connections.mov")

# Read frame by frame
while video.isOpened():
    ret, frame = video.read()
    
    # Check if the frame was read successfully
    if not ret:
        break  # Exit the loop if no frame is read

    # Convert BGR to HSV color space
    videoHSV = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    # Define the upper range and the lower range of the blue color
    lower_blue = np.array([110, 50, 50])
    upper_blue = np.array([130, 255, 255])
    
    # Threshold the HSV image for the range of blue color
    maskVideo = cv2.inRange(videoHSV, lower_blue, upper_blue)
    
    # Extract the blue color object alone
    result = cv2.bitwise_and(frame, frame, mask=maskVideo)

    # hsv video
    originalVideo = cv2.resize(videoHSV, (400, 400))
    cv2.imshow('HSVVideo', originalVideo)
    
    # Resize and display the original video
    originalVideo = cv2.resize(frame, (400, 400))
    cv2.imshow('OriginalVideo', originalVideo)
    
    # Resize and display the mask video
    maskVideo = cv2.resize(maskVideo, (400, 400))
    cv2.imshow('maskVideo', maskVideo)
    
    # Resize and display the result
    finalVideo = cv2.resize(result, (400, 400))
    cv2.imshow('finalVideo', finalVideo)
    
    if cv2.waitKey(1) == ord('q'):
        break
    
video.release()
cv2.destroyAllWindows()
```
