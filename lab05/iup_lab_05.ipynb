```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Negative transformation for color images

img = cv.imread("D:\\SLIIT\\4th_year\\2nd semester\\IUP\LABS\\lab05\\apples.jpg", 1)

# Apply negative transformation for color image: ----------------------------------
# 1st Method
# negative_img = 255 - img
# 2nd Method
negative_img = cv.bitwise_not(img)

# --------------------------------------------------------------------------

plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.show()

plt.imshow(cv.cvtColor(negative_img, cv.COLOR_BGR2RGB))
plt.show()
```

```python
# Power-law transformation to improve the contrast of a dark image

img = cv.imread("D:\\SLIIT\\4th_year\\2nd semester\\IUP\LABS\\lab05\\gamma.jpg", 0)

# Apply 2.2 gamma value ------------------------------------------------
# gamma_2_2 = np.array(255*(img/255)**2.2, dtype='uint8')
gamma1 = 2.2

# Apply 0.4 gamma value
# gamma_0_4 = np.array(255*(img/255)**0.4, dtype='uint8')
gamma2 = 0.4

gamma_2_2 = np.power(img, gamma1)
gamma_0_4 = np.power(img, gamma2)

# --------------------------------------------------------------------------

# gamma_img = cv.hconcat([gamma_2_2, gamma_0_4])

plt.imshow(gamma_2_2, "gray")
plt.show()

plt.imshow(gamma_0_4, "gray")
plt.show()
```

```python
# Log transformation to improve the dynamic range of an image

img = cv.imread("D:\\SLIIT\\4th_year\\2nd semester\\IUP\LABS\\lab05\\log.jpg", 0)

# Apply log transformation -----------------------------------------------

# 1st Method
# c = 255 / np.log(1 + np.max(img))
# log_img = c * np.log( img + 1)

# 2nd Method
log_img = (np.log(img + 1 ) / (np.log( 1 + np.max( img )))) * 255

# Ignore the error
np.seterr(divide='ignore')

# --------------------------------------------------------------------------

log_img = np.array(log_img, dtype= np.uint8)

plt.imshow(img, cmap="gray")
plt.show()

plt.imshow(log_img, cmap="gray")
plt.show()
```
