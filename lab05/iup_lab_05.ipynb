```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Negative transformation for color images

img = cv.imread("D:\\SLIIT\\4th_year\\2nd semester\\IUP\LABS\\lab05\\apples.jpg", 1)

# Apply negative transformation for color image: ----------------------------------
# 1st Method
# negative_img = 255 - img
# 2nd Method
negative_img = cv.bitwise_not(img)

# --------------------------------------------------------------------------

plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.show()

plt.imshow(cv.cvtColor(negative_img, cv.COLOR_BGR2RGB))
plt.show()
```

```python
# Power-law transformation to improve the contrast of a dark image

img = cv.imread("D:\\SLIIT\\4th_year\\2nd semester\\IUP\LABS\\lab05\\gamma.jpg", 0)

# Apply 2.2 gamma value ------------------------------------------------
# gamma_2_2 = np.array(255*(img/255)**2.2, dtype='uint8')
gamma1 = 2.2

# Apply 0.4 gamma value
# gamma_0_4 = np.array(255*(img/255)**0.4, dtype='uint8')
gamma2 = 0.4

gamma_2_2 = np.power(img, gamma1)
gamma_0_4 = np.power(img, gamma2)

# --------------------------------------------------------------------------

# gamma_img = cv.hconcat([gamma_2_2, gamma_0_4])

plt.imshow(gamma_2_2, "gray")
plt.show()

plt.imshow(gamma_0_4, "gray")
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load an image in grayscale
img = cv.imread("D:\\SLIIT\\4th_year\\2nd semester\\IUP\\LABS\\lab05\\log.jpg", 0)

if img is None:
    print("Error loading image")
else:
    # Apply log transformation -----------------------------------------------

    # Using the second method
    img_float = img.astype(float)  # Convert to float for precision in log calculation
    log_img = (np.log(img_float + 1) / np.log(1 + np.max(img_float))) * 255

    # Handle division by zero and other errors silently
    np.seterr(divide='ignore', invalid='ignore')

    # Clip the transformed values to be in the proper range and convert to uint8
    log_img = np.clip(log_img, 0, 255).astype(np.uint8)

    # --------------------------------------------------------------------------

    # Show the original image
    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.title("Original Image")
    plt.imshow(img, cmap="gray")
    plt.axis('off')

    # Show the log transformed image
    plt.subplot(122)
    plt.title("Log Transformed Image")
    plt.imshow(log_img, cmap="gray")
    plt.axis('off')

    plt.show()
```
