```python

img = cv2.imread('D:/UNI_FILEs/YEAR_4/Y4S2/IUP/Week 1 - 6 February - 12 February/Lab_01/flower.jpg', 1)
cv2.imshow('Original Image', img)

imgGray = cv2.imread('D:/UNI_FILEs/YEAR_4/Y4S2/IUP/Week 1 - 6 February - 12 February/Lab_01/flower.jpg', 0)
cv2.imshow('Gray Image', imgGray)

res, binary = cv2.threshold(imgGray, 120, 255, cv2.THRESH_BINARY)
cv2.imshow('Binary Image', binary)

print(res)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
fig,ax = plt.subplots(1,3, figsize = (15, 15))
ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax[0].axis('on')
ax[1].imshow(cv2.cvtColor(imgGray, cv2.COLOR_BGR2RGB))
ax[1].axis('off')
ax[2].imshow(cv2.cvtColor(binary, cv2.COLOR_BGR2RGB))
ax[2].axis('on')
```
